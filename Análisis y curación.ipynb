```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd
import seaborn
seaborn.set_context('talk')
from sqlalchemy import create_engine, text
import plotly
from sqlalchemy import create_engine

url = "https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/melb_data.csv"

melb_df = pd.read_csv(url)
pd.options.display.float_format = '{:.2f}'.format
melb_df[:10]

```

```python
#recuento de las filas de cada suburb
melb_df['Suburb'].value_counts()
```

```python
#Crear base de datos en SQLite para realizar una tabla
BD = create_engine('sqlite:///melb_data.sqlite3', echo=True)

#Creo una tabla con la base anterior
melb_df.to_sql('propiedades', con = BD, if_exists="replace")
```

```python
from sqlalchemy.future.engine import Connection
#Cantidad de registros totales por ciudad
#No entiendo bien que hace esto
query1 = "SELECT Suburb FROM propiedades WHERE Suburb GROUP BY Suburb"
query2 = "SELECT CouncilArea FROM propiedades WHERE CouncilArea GROUP BY CouncilArea"

query8 = """
SELECT Suburb, COUNT(Suburb)
FROM propiedades 
WHERE Suburb
GROUP BY Suburb
ORDER BY AVG(Suburb) DESC
"""
query9 = """
SELECT CouncilArea, COUNT(Suburb)
FROM propiedades 
WHERE CouncilArea
GROUP BY CouncilArea
"""
queries = [query1, query2]
Queries = [query8, query9]
print("--------")

#with BD.connect() as con:
#      rs = con.execute(query2)
#      for row in rs:
#          print(row)
#print("--------")

with BD.connect() as con:
    for query in Queries:
      rs = con.execute(query)
      print(query)
      for row in rs:
          print(row)

      print('\n\n')
```
